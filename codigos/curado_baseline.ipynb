```
import pickle
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```
with open('/content/drive/My Drive/ICFernando/SEER Remodelado/bases/curado/X_train.pkl', 'rb') as arq:
    X_train = pickle.load(arq)
with open('/content/drive/My Drive/ICFernando/SEER Remodelado/bases/curado/X_valid.pkl', 'rb') as arq:
    X_valid = pickle.load(arq)
with open('/content/drive/My Drive/ICFernando/SEER Remodelado/bases/curado/X_test.pkl', 'rb') as arq:
    X_test = pickle.load(arq)
with open('/content/drive/My Drive/ICFernando/SEER Remodelado/bases/curado/y_train.pkl', 'rb') as arq:
    y_train = pickle.load(arq)
with open('/content/drive/My Drive/ICFernando/SEER Remodelado/bases/curado/y_valid.pkl', 'rb') as arq:
    y_valid = pickle.load(arq)
with open('/content/drive/My Drive/ICFernando/SEER Remodelado/bases/curado/y_test.pkl', 'rb') as arq:
    y_test = pickle.load(arq)
```

```
%cd codigos
from custom_metrics import Metricas
```

```
ml = LogisticRegression(penalty='l2', dual=False, tol=0.0001, C=1.0, fit_intercept=True, intercept_scaling=1,
                        class_weight=None, random_state=1, solver='lbfgs', max_iter=1000,
                        multi_class='auto', verbose=0, warm_start=False, n_jobs=None, l1_ratio=None)
ml.fit(X_train, y_train)
Metricas(y_test, ml.predict(X_test),  ml.predict_proba(X_test), nm_classes=['NAO_CURADO', 'CURADO'])
```

```
ml = GaussianNB()
ml.fit(X_train, y_train)
Metricas(y_test, ml.predict(X_test),  ml.predict_proba(X_test), nm_classes=['NAO_CURADO', 'CURADO'])
```

```
ml = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=10, min_samples_split=2, min_samples_leaf=1,
                            min_weight_fraction_leaf=0.0, max_features=None,
                            random_state=1, max_leaf_nodes=None, min_impurity_decrease=0.0,
                            ccp_alpha=0.001)
ml.fit(X_train, y_train)
Metricas(y_test, ml.predict(X_test),  ml.predict_proba(X_test), nm_classes=['NAO_CURADO', 'CURADO'])
```
