```python
import os
import time
import pandas as pd
import cudf
import glob
pd.set_option('display.max_columns', 200)
pd.set_option('display.max_rows', 100)
```

```python
home_path = '/home/svf'
base_tratada = 'base_col_sep_quimo'
nome_base = 'SEER_1975_2016_CUSTOM_TEXTDATA'
nome_dicionario = 'dicionario_dados_quimo.txt'
```

```python
def ler_dicionario(dic_var):
    lista = []
    for i in range(len(dic_var)):
        tam_var = dic_var.iloc[i].to_string()[41:43].strip()
        if tam_var[-1] == '.':
            tam_var = tam_var[0]
        nome_var = dic_var.iloc[i].to_string()[15:35].strip()
        pos_var = dic_var.iloc[i].to_string()[11:14].strip()
        pos_var = int(pos_var)
        tam_var = int(tam_var)
        lista.append([nome_var, pos_var, tam_var])
        lista_nomes = [nome for nome, pos, tam in lista]
    return lista, lista_nomes
```

```python
dic_var = pd.read_csv(os.path.join(home_path, 'config', 'dicionario_dados_quimo.txt'), header=None)
```

```python
base_paths = os.path.join(home_path, nome_base, 'incidence')
base_tratada = os.path.join(home_path, base_tratada)
dic_var = pd.read_csv(os.path.join(home_path, 'config', nome_dicionario), header=None)
lista, lista_nomes = ler_dicionario(dic_var)
```

```python
for i, base_path in enumerate(glob.iglob(base_paths + '/**/*.TXT', recursive=True)):
    antigo = time.time()
    db_sep_path = os.path.join(base_tratada, base_path.split(sep='/')[-2], base_path.split(sep='/')[-1][:-4] + '.csv')
    if not os.path.exists(os.path.dirname(db_sep_path)):
        os.makedirs(os.path.dirname(db_sep_path))
    db = cudf.read_csv(base_path, header=None)
    for nome, pos, tam in lista:
        pos -= 1
        db[nome] = db['0'].str.slice(pos, pos + tam).str.strip()
    db = db.drop('0', axis=1)
    db.to_csv(db_sep_path, index=False)
    print('Tempo: {0} s | Salvo em: {1}'.format(round(time.time() - antigo, 2), db_sep_path))
```
